Возьмем отметку 0,9 как оптимальное покрытие, так как 1 несет в себе риски простоев машин на линии, а разница в 10% с высокой долей вероятности может быть покрыта отказами от поездок и неодновременностью вызовов.

Рассчитаем, сколько поездок необходимо выполнить находящимся на линии автомобилям, чтобы обеспечить покрытие 0,9.

**$\color{blue}{\text{КОММЕНТАРИЙ}}$**

Великолепно что ты аргументируешь выбор оптимального покрытия

```python
# переведем значение покрытия в проценты
df_general_aggregation['estimated_coverage'] = (df_general_aggregation['estimated_coverage'] * 100).astype('int')
display(df_general_aggregation.head())
```

**$\color{blue}{\text{КОММЕНТАРИЙ}}$**

Наверное нам необяхательно умножать столбец на 100 Просто взять для расчетов в качестве оптимального покрытия не 90 а 0,9

```python
# рассчитаем число поездок, обеспечивающее покрытие 90%
df_general_aggregation['optimal_trips_cnt'] = 0
for index, row in df_general_aggregation.iterrows():
    if row['estimated_coverage'] < 90:
        optimal_trips_cnt = row['trips'] * 90 / row['estimated_coverage']
    else:
        optimal_trips_cnt = row['trips']
    df_general_aggregation.at[index, 'optimal_trips_cnt'] = round(optimal_trips_cnt)
df_general_aggregation.head()
```

**$\color{blue}{\text{КОММЕНТАРИЙ}}$**

Название столбца optimal_coverage немного не соответствует его содержанию Скорее здесь у нас количество поездок при оптимальном покрытии

<div class="alert alert-info"> <b>Комментарий студента:</b> обновлено.</div>

```python
# рассчитаем число машин, способных выполнить необходимое число поездок
df_general_aggregation['cars_optimal'] = (df_general_aggregation['optimal_trips_cnt'] * df_general_aggregation['cars'] / df_general_aggregation['trips']).round()
df_general_aggregation['cars_difference'] = (df_general_aggregation['cars_optimal'] - df_general_aggregation['cars']).astype('int')
df_general_aggregation.head()
```
